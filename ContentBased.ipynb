```python
import numpy as np 
import pandas as pd 
import ast
```

```python
credits = pd.read_csv('./tmdb_5000_credits.csv') 
movies = pd.read_csv('./tmdb_5000_movies.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
MoviesData = movies.merge(credits,on='title')
```

```python
MoviesData.head(1)
```

```python
MoviesData = MoviesData[['movie_id','title','overview','genres','keywords','cast','crew']]
MoviesData.head(1)
```

```python
MoviesData.dropna(inplace=True)  # will drop all NAN vales 
MoviesData.head()
```

```python
 # this will extract name and convert the data inside genre col into an array of only genre names 
 # this function will work the same for keywords
def convert_to_names_only(col):
    List = []
    for i in ast.literal_eval(col):
        List.append(i['name']) 
    return List

MoviesData['genres'] = MoviesData['genres'].apply(convert_to_names_only)  
MoviesData['keywords'] = MoviesData['keywords'].apply(convert_to_names_only)

```

```python

ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert_to_top5_names(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 5: # selecting only 5 actors now
            L.append(i['name'])
        counter+=1
    return L 
```

```python
MoviesData['cast'] = MoviesData['cast'].apply(convert_to_top5_names)
MoviesData['cast'] = MoviesData['cast'].apply(lambda x:x[0:5])
MoviesData.head()
```

```python
def Get_director_only(text):
    List = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            List.append(i['name'])
    return List
```

```python
MoviesData['crew'] = MoviesData['crew'].apply(Get_director_only)
```

```python
def Remove_spaces(L):    # replacing spaces with no spaces 
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
MoviesData['cast'] = MoviesData['cast'].apply(Remove_spaces)
MoviesData['crew'] = MoviesData['crew'].apply(Remove_spaces)
MoviesData['genres'] = MoviesData['genres'].apply(Remove_spaces)
MoviesData['keywords'] = MoviesData['keywords'].apply(Remove_spaces)
```

```python
MoviesData.head()
```

```python
#seperating each word in the overview as later we will use the word technique to find most repaeated words
MoviesData['overview'] = MoviesData['overview'].apply(lambda x:x.split()) 

```

```python
MoviesData.head()
```

```python
# will create another col for tags by appending all the features into it
MoviesData['tags'] = MoviesData['overview'] + MoviesData['genres'] + MoviesData['keywords'] + MoviesData['cast'] + MoviesData['crew']
```

```python
new_Movie_Data = MoviesData.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new_Movie_Data['tags'] = new_Movie_Data['tags'].apply(lambda x: " ".join(x))
```

```python
new_Movie_Data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000,stop_words='english') # select 5000 words for a vector, donot select the stop words (a, in, the)
    
```

```python
vector = cv.fit_transform(new_Movie_Data['tags']).toarray() # creating vector using the tag
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)  # now we wll apply the cosine similarity rule to find the angular distances between eah vetor 
```

```python
similarity
```

```python
def recommend_movie(movie):
    index = new_Movie_Data[new_Movie_Data['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_Movie_Data.iloc[i[0]].title)
```

```python
recommend_movie('Batman')
```
